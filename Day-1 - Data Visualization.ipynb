```python
x=np.arange(0,10)
y=x*x

plt.title("1st Assignment")
plt.xlabel("X-Axis data")
plt.ylabel("Y-Axis data")
plt.plot(x,y,'g*--',linewidth=1,markersize=10)
```

#### Assignment-2

```python
days = [1,2,3,4,5,6,7]
sales_one = [160, 150, 140, 145, 175, 165, 180]
sales_two = [70, 90, 160, 150, 140, 145, 175]

plt.figure(figsize=(6,5), dpi=150)


plt.plot(days, sales_one,'g*--',linewidth=1,markersize=10)
plt.plot(days, sales_two,'r*--',linewidth=1,markersize=10)

plt.xlabel('days of the week')
plt.ylabel('sales')
plt.grid()
plt.show()
```
