##### Q) Plot using sns.relplot, line plot, box plot on fmri data set

```python
# Import Libraries
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df_fmri = sns.load_dataset("fmri")
df_fmri.head()
```

```python
df_fmri.info()
```

```python
df_fmri.describe()
```

```python
df_fmri.subject.value_counts()
```

```python
df_fmri.event.value_counts()
```

```python
df_fmri.region.value_counts()
```

### relplot

```python
# relplot - by default it takes scatterplot

sns.relplot(x='signal', 
            y='timepoint', 
            data=df_fmri);
```

```python
#above diagram and below diagram both are same

sns.relplot(x='signal', 
            y='timepoint', 
            data=df_fmri, 
            kind='scatter');
```

```python
sns.relplot(x='signal', 
            y='timepoint', 
            data=df_fmri, 
            kind='scatter',
            size='subject');
```

```python
sns.relplot(x='signal', 
            y='timepoint', 
            data=df_fmri, 
            kind='scatter',
            size='subject',
            hue='event');
```

```python
hue_colors = {'parietal': 'red', 
              'frontal': 'blue'}

sns.relplot(x='signal', 
            y='timepoint', 
            data=df_fmri, 
            kind='scatter',
            size='subject',
            hue='region',
            palette=hue_colors);
```

### Line Plot

```python
sns.lineplot(data=df_fmri, x="signal", y="timepoint")
```

```python
sns.lineplot(data=df_fmri, x="signal", y="timepoint", hue="subject", palette="rainbow")
```

```python
sns.lineplot(data=df_fmri, x="signal", y="timepoint", hue="event", palette="rainbow")
```

```python
sns.lineplot(
    data=df_fmri, x="signal", y="timepoint", hue="event", style="region",palette="rainbow"
)
```

### Boxplot

```python
sns.boxplot(df_fmri["signal"])
```

```python
sns.boxplot(data = df_fmri, x = "signal", y = "timepoint", hue = "event")
```

```python
sns.boxplot(data =df_fmri,x="timepoint",y ="signal",hue ="event")
```

```python
sns.boxplot(df_fmri["timepoint"])
plt.savefig("BoxPlot.jpg")
```

```python
sns.boxplot(data =df_fmri,x="timepoint",y ="signal",hue ="subject")
plt.savefig("BoxPlot.jpg")
```

```python
sns.boxplot(data =df_fmri,x="timepoint",y ="signal",hue ="region")
plt.savefig("BoxPlot.jpg")
```
